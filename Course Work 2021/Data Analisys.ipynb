```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_colwidth',None)
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('vyborka.csv', decimal = '.', sep = ',', error_bad_lines=False)
```

```python
df
```

```python
df.notna().describe()
```

```python
numeric = df.dtypes[df.dtypes == 'object'].index
numeric
```

```python
df.describe()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(15,14))

plt.title('Pearson Correlation of Variables', y=1.05, size=12)
sns.heatmap(df[['LNHOURWAGE', 'WORK', 'AGE', 'AGE2','EDUC0','EDUC1', 'EDUC2', 'EDUC3', 'EDUC4', 'FEMALE', 'SKILLS', 'NKIDS', 'OBL_CENTRE', 
                 'PGT_SELO', 'OCCUP0', 'OCCUP1', 'IND0', 'IND1', 'IND2', 'IND3']].corr(),
            linewidths=0.1, vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.scatterplot(y = 'LNHOURWAGE', x = 'AGE', data=df, palette = 'flare')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.kdeplot(x = 'LNHOURWAGE', data=df, palette = 'Set3')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.kdeplot(data=df, x="LNHOURWAGE", hue="WAVE", palette = 'flare')
```

```python
grouped = pd.pivot_table(data=df,index=['WAVE'], values=['LNHOURWAGE'])
grouped
```

```python
sns.lineplot(data = grouped, palette = 'flare')
_ = plt.title('Mean LNHOURWAGE')
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.histplot(data=df, x="LNHOURWAGE", hue="MARRIED", palette = 'flare', fill = True)
```

```python
grouped = pd.pivot_table(data=df,index=['MARRIED'], values=['LNHOURWAGE'])
grouped
```

```python
grouped = pd.pivot_table(data=df,index=['EDUC'], values=['LNHOURWAGE'])
grouped
```

```python
ED = pd.DataFrame([0, 1, 2, 3, 4], columns = ['EDUC'])
grouped = pd.concat([grouped, ED], axis = 1)
grouped
```

```python
sns.barplot(x = 'EDUC', y = 'LNHOURWAGE', data=grouped, palette = 'flare')
sns.lineplot(x = 'EDUC', y = 'LNHOURWAGE', data=grouped, palette = 'flare')
_ = plt.title('Mean LNHOURWAGE value, devided by EDUC')
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.catplot(
    data=df,
    x="AGE", y='LNHOURWAGE', col='EDUC', col_wrap=3, margin_titles = True, palette = 'flare'
)
_ = plt.title('Scatter plot in axis AGE - LNHOURWAGE, devided by EDUC')
```
